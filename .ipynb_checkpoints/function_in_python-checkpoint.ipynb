# Functions in Python

### definition:
### a function is a block of code that performes a specific task.
###  functions help in organizing code, reusing coe, and improving readiblity.

```python
## really good syntax for functions
def function_name(parameters):
    """Docstring"""  ## Docstring is used to write some kinds of comments
    # Function body
    return expression  ## return value is not necessary always
```

```python
## why do we require functions?

num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
## use function for reusing a code

def even_or_odd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("the number is even")
    else:
        print("the numbers is odd")
    
```

```python
## call the function

even_or_odd(35)
```

```python
## function with multiple parameters

def add(a,b):
    return a + b

result = add(2,4)
print(result)    
```

```python
## default parameters

def greet(name):
    print(f"Hello {name}, welkom to the paradijs")

greet("Krish")
```

```python
def greet(name = "guest"):
    print(f"Hello {name}, welkom to the paradijs")

greet()
```

```python
def greet(name = "guest"):
    print(f"Hello {name}, welkom to the paradijs")

greet("Ali")
```

```python
### variable length arguments

## positional and keywords arguments

def print_numbers(*krish):  ## args stands for arguments. we have a lot of arguments it can be taken any name like "krish"
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,"ali", 7.8, "krish")  ## thse parameters we call positional arguments
```

```python
## positional and keywords arguments

def print_numbers(*args):  ## args stands for arguments. we have a lot of arguments it can be taken any name like "krish" best practice is args
    for number in args:
        print(number)  
```

```python
print_numbers(1,2,3,4,"ali", 7.8, "krish")  
```

```python
### keywords arguments
### in keywords arguments all the parameters should be in the form of key value pairs
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    
```

```python
print_details(name = "alireza", age = 42, hight = 183, city = "Rotterdam", position = "data analyst")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4, "Ali", name = "alireza", age = 42, hight = 183, city = "Rotterdam", position = "data analyst") 
```

# Python functions examples 

### Example 1: Temperature Conversion

```python
def convert_temperature(temp, unit):
    """ This function converts temperature between celsius and fahrenheit"""
    if unit == "C":
        return temp * (9/5) + 32  ## celuis to fahrenheit
    elif unit == "F":
        return (temp - 32)*5/9 ## ferenheit to celcius
    else:
        return None

print(convert_temperature(25, "C"))
print(convert_temperature(26, "F"))   
    
```

## password streng checker

```python
def is_strong_password(password):
    """ This function checks if the password is strong or not"""
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    else:
        return True
        
## calling the function

print(is_strong_password("weakPwd"))

print(is_strong_password("weakPwd0)"))      
print(is_strong_password("weakPwdee0"))    
```

## Lambda Functions in Python
#### Lambda functions are small anonymous(without name) functions defined using the lambda keyword. They can have any number of arguments but only one expression. (one expression means one logic) 
#### They are commonly used for short operations or as arguments to higher-order functions.

```python
#Syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition = lambda a,b: a+b
print(type(addition))
print(addition(5,6))
```

```python
def even(num):
    if num%2 ==0:
        return True

even(26)
```

```python
lambda num: num%2 == 0
even(24)
```

```python
def addition(x,y,z):
    return x + y +z

addition(2,3,5)
```

```python
addition1 = lambda x,y,z: x+y+z

addition1(4,6,8)
```

```python
## map function and filter function

## map() function applies a function to all items in a list

numbers = [1,2,3,4,5,6]
def square(number):
    return number**2
    
square(2)
```

```python
list(map(lambda x:x**2, numbers))
```
