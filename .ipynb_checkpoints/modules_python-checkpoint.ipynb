## Moudules in Python

#### In Python, modules and packages help organize and reuse code. 
#### Here's a comprehensive guide on how to import them.

```python
import math
```

```python
math.sqrt(16)
```

```python
from math import sqrt,pi

print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np

np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
import numpy as np
print(np.__version__)

```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import* 
sqrt(16)
```

```python
import sys
print(sys.executable)
```
