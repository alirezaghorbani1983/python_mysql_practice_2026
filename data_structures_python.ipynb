### Create a To Do List To Keep Track OF Tasks

```python
to_do_list = ["buy groceris", "make an exercise", "do homework", "clean the room", "pay bills"]

## adding the task

to_do_list.append("schedule meetings")
to_do_list.append("go for a run")

```

```python
## Create a list to store and calculate average grades for students
grades = [87, 69, 59, 81]

# adding a new grade 
grades.append(95)

average_grade = sum(grades)/len(grades)
print(f"Average grade: {average_grade:.2f}")
```

```python
## find the highest and lowest grade

highest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

### iterating over dictionaries



```python
## you can use loops to iterrate over dictionries, keys, values, items

## iterating over keys

for keys in student.keys():
    print(keys)
```

```python
## iterating over values

for values in student.values():
    print(values)
```

## nested dictionaries

```python
## how create nested dictionaries

students = {
    "student1":{"name":"krish", "age":32},
    "student2":{"name":"krish", "age":35}
}

print(students)
```

```python
## accessing nested dictionary elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## iterating over nested dictionary

for student_id, student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key, value in student_info.items():
        print(f"{key} : {value}")
    
```

### dictionary comprehension



```python
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dictionary comprehension

evens = {x:x**2 for x in range(10) if x%2 ==0}
print(evens)
```

### practical examples

```python
## use a dictionary to count the frequency of elemets in list

numbers = [1,2,2,3,3,3,4,4,4,4]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] +=1
    else:
        frequency[number] = 1
print(frequency)
```

```python
## merge 2 dictionaries into one

dict1 = {"a":1, "b":2}
dict2 = {"b":3, "c":4}

merged_dict = {**dict1, **dict2}

print(merged_dict)
```
