## Real-World Examples Using Lists in Python

### Create a To Do List To Keep Track OF Tasks

```python
to_do_list = ["buy groceris", "make an exercise", "do homework", "clean the room", "pay bills"]

## adding the task

to_do_list.append("schedule meetings")
to_do_list.append("go for a run")

```

```python
to_do_list
```

```python
# removing a completed task

to_do_list.remove("clean the room")

to_do_list
```

```python
## checking if a task is in the list

if "pay bills" in to_do_list:
    print("don't forget to pay the utility bills")
```

```python
print("to do list remaing")

for task in to_do_list:
    print(f"-{task}")
```

```python
## Create a list to store and calculate average grades for students
grades = [87, 69, 59, 81]

# adding a new grade 
grades.append(95)

average_grade = sum(grades)/len(grades)
print(f"Average grade: {average_grade:.2f}")
```

```python
## find the highest and lowest grade

highest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

## managing an inventory

```python
## Use a list to manage inventory items in a store
```

```python
## managing an inventory

inventory = ["apples", "bananas", "oranges", "grapes"]

# adding a new item 
inventory.append("strawberries")

## removing an item that is out of stock
inventory.remove("bananas")
```

```python
## checking an item is in stock

item = "strawberries"

if item in inventory:
    print(f"{item} are in stock")
else:
    print(f"{item} is out of stock")
```

```python
## printing the inventory list

print("Inventory list: ")

for item in inventory:
    print(f"-{item}")
```

```python
feedback = ["great service!", "very satisfied", "Could be better", "Excellent experience"]

## adding new feedback
feedback.append("not happy with the service")


print(feedback)
```

```python
## counting specific feedback

positive_feedback_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())

print(f"positive feedback count: {positive_feedback_count}")
```

```python
## printing all feedback

print("user feedback: ")
for comment in feedback:
    print(f"-{comment}")
```

```python
## mathematical operations

set1 = {1,2,3,4,6}
set2 = {4,5,6,7,8,9}

## Union  == > union method combines everything

union_set = set1.union(set2)
print(union_set)

```

```python
## intersectional

## mathematical operations

set1 = {1,2,3,4,6}
set2 = {4,5,6,7,8,9}

## Union  == > union method combines everything

intersection_set = set1.intersection(set2)
print(intersection_set)

print(set1)
```

```python
set1.intersection_update(set2)
print(set1)
```

```python
## difference
set1 = {1,2,3,4,6}
set2 = {4,5,6,7,8,9}
set1.difference(set2)
```

```python
set1.difference_update(set2)
print(set1)
```
