## File Operation- Read And Write Files


#### File handling is a crucial part of any programming language. 
#### Python provides built-in functions and methods to read from and write to files, both text and binary. 
#### This lesson will cover the basics of file handling, including reading and writing text files and binary files.

#### read a file

```python
## read a whole file

with open('example.txt', mode = 'r') as file:
    for line in file:
        content = file.read()
        print(content)
        
                                    
```

```python
## read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())  ## strip() eliminates extra space between lines(strip() removes the newline character
        
```

### writing a file (with overwriting)
#### overwriting -->  replacing the previous content of a file with new content

```python
### write is a method that help us to write anything inside our file
### with this code would be overritten entire content of previous file ie, example.txt
with open('example.txt', 'w') as file:
    file.write("Hello World!\n")
    file.write('this is a new line')
```

### write a file (without overwriting)

```python
with open('example.txt', 'a') as file: ### a is append mode
    file.write(" Append operation taking place!\n")
    
```

```python
### writing a list of lines to a file

lines = ['first line \n', 'second line \n', 'third line\n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### respect to binary files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```
