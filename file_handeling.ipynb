## File Operation- Read And Write Files


#### File handling is a crucial part of any programming language. 
#### Python provides built-in functions and methods to read from and write to files, both text and binary. 
#### This lesson will cover the basics of file handling, including reading and writing text files and binary files.

#### read a file

```python
## read a whole file

with open('example.txt', mode = 'r') as file:
    for line in file:
        content = file.read()
        print(content)
        
                                    
```

```python
## read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())  ## strip() eliminates extra space between lines(strip() removes the newline character
        
```

### writing a file (with overwriting)
#### overwriting -->  replacing the previous content of a file with new content

```python
### write is a method that help us to write anything inside our file
### with this code would be overritten entire content of previous file ie, example.txt
with open('example.txt', 'w') as file:
    file.write("Hello World!\n")
    file.write('this is a new line')
```

### write a file (without overwriting)

```python
with open('example.txt', 'a') as file: ### a is append mode
    file.write(" Append operation taking place!\n")
    
```

```python
### writing a list of lines to a file

lines = ['first line \n', 'second line \n', 'third line\n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### respect to binary files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

### some practical example

```python
## read the content from source text file and write to a destination text of file

## first we read a file and put it in a variable : content
with open('example.txt', 'r') as source_file:
    content = source_file.read()

## then we write a new file (destination.txt)

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
    
```

```python
## Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file

def count_text_file(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_name = 'example.txt'
lines, words, characters = count_text_file(file_name)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
```

## Working With File Paths

When working with files in Python, handling file paths correctly is crucial to ensure your code works across different operating systems and environments. Python provides several modules and functions for working with file paths effectively.

```python
#### Using the os module
import os
cwd=os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## create a new directory
new_directory="package_2"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' create")
```

```python
## Listing Files And Directories
items=os.listdir('.')
print(items)
```

```python
### Joining Paths

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
dir_name="folder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
path='example1.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exists")
```

```python
#Checking if a Path is a File or Directory
import os

path = 'example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
```

```python
## Getting the absolute path
relative_path='example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
