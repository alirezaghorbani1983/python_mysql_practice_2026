# Functions in Python

### definition:
### a function is a block of code that performes a specific task.
###  functions help in organizing code, reusing coe, and improving readiblity.

```python
## really good syntax for functions
def function_name(parameters):
    """Docstring"""  ## Docstring is used to write some kinds of comments
    # Function body
    return expression  ## return value is not necessary always
```

```python
## why do we require functions?

num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
## use function for reusing a code

def even_or_odd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("the number is even")
    else:
        print("the numbers is odd")
    
```

```python
## call the function

even_or_odd(35)
```

```python
## function with multiple parameters

def add(a,b):
    return a + b

result = add(2,4)
print(result)    
```

```python
## default parameters

def greet(name):
    print(f"Hello {name}, welkom to the paradijs")

greet("Krish")
```

```python
def greet(name = "guest"):
    print(f"Hello {name}, welkom to the paradijs")

greet()
```

```python
def greet(name = "guest"):
    print(f"Hello {name}, welkom to the paradijs")

greet("Ali")
```

```python
### variable length arguments

## positional and keywords arguments

def print_numbers(*krish):  ## args stands for arguments. we have a lot of arguments it can be taken any name like "krish"
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,"ali", 7.8, "krish")  ## thse parameters we call positional arguments
```

```python
## positional and keywords arguments

def print_numbers(*args):  ## args stands for arguments. we have a lot of arguments it can be taken any name like "krish" best practice is args
    for number in args:
        print(number)  
```

```python
print_numbers(1,2,3,4,"ali", 7.8, "krish")  
```

```python
### keywords arguments
### in keywords arguments all the parameters should be in the form of key value pairs
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    
```

```python
print_details(name = "alireza", age = 42, hight = 183, city = "Rotterdam", position = "data analyst")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4, "Ali", name = "alireza", age = 42, hight = 183, city = "Rotterdam", position = "data analyst") 
```
