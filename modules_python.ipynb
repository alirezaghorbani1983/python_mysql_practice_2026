### data serialization

```python
import json
data={'name': 'alireza', 'age': 4}  ## data is a dictionary

json_str = json.dumps(data)  ## dump() ==> returnes us a json format from a dictionary

print(json_str)  ## json has always double quotation
print(type(json_str))

parsed_json = json.loads(json_str)  ## loads() ==> parse_json is a module or object.
print(parsed_json)
print(type(parsed_json))
### The loads() function inside it parses a JSON string and converts it into a usable Python structure, usually a dictionary.
```

### make a csv file with open function

#### The open() function opens a file.  with open('example.csv', mode='w', newline='') as file:
#### It opens the file example.csv   With mode 'w', which means write mode  
####  With mode 'w', which means write mode


```python
#### writer is a CSV writer object.
### writer is a CSV writer object.
```

#### writer.writerow(['Ali', 33]) writes one row into the CSV file

```python
## csv library
## make a csv file

import csv
with open('example.csv', mode ='w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Ali', 33])

## this piece of code creates a csv file named example.csv

with open('example.csv', mode = 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
