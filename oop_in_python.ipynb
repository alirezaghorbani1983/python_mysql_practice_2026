```python
### instance variable and methodes 
### instance variable
class Dog:
    ## constructor
    ## define instance variable (define attributes)
    def __init__(self,name, age):
        self.name = name
        self.age = age

## create an object
dog1 = Dog('buddy',3)
print(dog1)
print(dog1.name)
print(dog1.age)
    
    
```

```python
dog2 = Dog('tito', 4)
print(dog2.name)
print(dog2.age)
```
